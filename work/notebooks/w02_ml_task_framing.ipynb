## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*


I chose **Structured Content Archetype Clustering**.

This is an **unsupervised clustering** problem.

Unlike a classification task, there are no predefined labels indicating which type of content each page belongs to. Instead, the objective is to discover natural groups of content pages that exhibit similar search performance, engagement, freshness, and content characteristics.

The discovered clusters will serve as **content archetypes** that help content teams understand their inventory and apply different management strategies for different groups of pages.


## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

This lane does **not** use a predefined target variable because clustering is an unsupervised learning task.

Instead of predicting an existing label, the objective is to discover groups of pages that share similar observable characteristics.

The clustering process will use variables such as:

- impressions
- clicks
- sessions
- CTR
- average position
- engagement rate
- content age
- freshness
- word count

These observable signals will be used to identify recurring performance archetypes.

The resulting cluster assignments are outputs of the model rather than labels provided by the dataset.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Because this is an unsupervised clustering task, traditional classification metrics such as accuracy or ROC AUC are not applicable.

The primary quantitative evaluation metric will be the **Silhouette Score**, which measures how well-separated and internally consistent the discovered clusters are.

However, numerical quality alone is not sufficient.

The clusters should also be interpretable and useful for business decisions. Each cluster should represent a meaningful content archetype with distinct characteristics and a clear recommended action.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

Each row in the dataframe represents **one content page**.

Each content page contains observable search, engagement, freshness, and content-related metrics aggregated over a 90-day period.

The clustering algorithm will assign every page to one archetype based on these observable characteristics.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can only classify pages using manually defined thresholds, such as "high impressions" or "low CTR." However, content performance is influenced by many interacting variables, including search visibility, engagement, freshness, content length, and search intent.

These relationships are often multidimensional and difficult to capture using simple if-statements.

Clustering algorithms can identify recurring patterns across multiple variables simultaneously, allowing similar pages to be grouped together even when no obvious rule exists.

The resulting archetypes provide a richer understanding of the content inventory and support more targeted content management strategies.
